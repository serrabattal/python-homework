```python
import os
import csv
cwdPath = os.getcwd()
print(cwdPath)

csv_file = os.path.join("OneDrive\Desktop", "budget_data.csv")
print(csv_file)

# Setup variables

month_count = 0
total_profit = 0
starting_profit = 0
current_profit = 0
profit_change = 0
total_profit_change = 0
profit_average_change = 0

profit_list = []
rows = []
months = []
profit_changes = []

# Open the CSV file
with open("budget_data.csv", newline="") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=",")
    csv_header = next(csv_reader)
    
    # Count rows
    for row in csv_reader:
        # Get month count
        month_count = month_count + 1
        
        # Get total_profit
        profit_list.append(row[1])
        total_profit = total_profit + int(row[1])
        
        # Get list of months
        months.append(row[0])
        
        # Get list of P/Ls
        rows.append(row[1])
        
        #
        print("Month: " + str(month_count) + " | " + row[0])
        print("Starting profit: $" + str(starting_profit))
        
        # Assign current row's profit value to current_profit
        current_profit = row[1]
        print("Current profit: $" + row[1])
        
        # Calculate change between current and starting profit
        if starting_profit != 0:
            profit_change = int(current_profit) - int(starting_profit)
            profit_changes.append(profit_change)
            print("Profit change: $" + str(profit_change))
            # total_profit_change = profit_change + profit_change
            total_profit_change = total_profit_change + profit_change
            print("Total profit change: $" + str(total_profit_change))
        else:
            profit_change = 0
            profit_changes.append(profit_change)
            print("Profit change: $0")
        
        # Assign current profit to starting profit of next loop iteration
        starting_profit = current_profit
        print("New starting profit is $" + str(starting_profit))
        
        # Print breakline
        print("---------------------------------")
        
        # print(rows[month_count])
        # print("change: " + row[1])
        # starting_profit = row[1]
    
    # Print header
    print("Financial Analysis")
    print("---------------------------------")
    
    # 1. Print month_count
    print("Total Months: " + str(month_count))
    
    # 2. Print profit_total
    print("Total profit: $" + str(total_profit))
    
    # 3. Print average_change
    profit_average_change = total_profit_change / (month_count - 1)
    print("Profit change average: $" + str(profit_average_change))
    
    # 4. Get max and min changes by month
    max_profit_increase = max(profit_changes)
    max_month_increase = months[profit_changes.index(max_profit_increase)]
    print("Greatest Increase in Profits: " + str(max_month_increase) + " ($" + str(max_profit_increase) + ")")
    
    min_profit_increase = min(profit_changes)
    min_month_increase = months[profit_changes.index(min_profit_increase)]
    print("Greatest Decrease in Profits: " + str(min_month_increase) + " ($" + str(min_profit_increase) + ")")
    
    # min_profit_increase = min(profit_changes)
    # min_month_increase = months[]
    # print("Greatest Decrease in Profits: " + min_month_increase + " (" + str(min_profit_increase))

```
